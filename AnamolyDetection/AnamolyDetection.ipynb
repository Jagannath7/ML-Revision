```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mean1, cov1 = [0,0], [[0,1], [1,0]]
data1 = np.random.multivariate_normal(mean1, cov1, 500)

mean2, cov2 = [7,7], [[0,1], [1,0]]
data2 = np.random.multivariate_normal(mean2, cov2, 500)

plt.scatter(data1[:, 0], data1[:, 1])
plt.scatter(data2[:, 0], data2[:, 1])
plt.show()
```

```python
data = np.vstack([data1, data2])
np.random.shuffle(data)
```

```python
from sklearn.covariance import EllipticEnvelope
```

```python
ad = EllipticEnvelope()
```

```python
ad.fit(data)
```

```python
ad.predict(data[:20])
```

```python
preds_ad = ad.predict(data)
```

```python
print(len(preds_ad))
```

```python
plt.scatter(data[:,0],data[:,1], c=preds_ad)
plt.show() #purple - outliers
```

```python
from sklearn.svm import OneClassSVM
```

```python
ocs = OneClassSVM()
```

```python
ocs.fit(data)
```

```python
preds_ocs = ocs.predict(data)
```

```python
plt.scatter(data[:,0],data[:,1], c=preds_ocs)
plt.show() #purple - outliers
```

```python
from sklearn.ensemble import IsolationForest
```

```python
ifr = IsolationForest()
```

```python
ifr.fit(data)
```

```python
preds_ifr = ifr.predict(data)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=preds_ifr)
plt.show()
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
lof = LocalOutlierFactor()
```

```python
preds_lof = lof.fit_predict(data)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=preds_lof)
plt.show()
```
