```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(n_samples=1000, factor=0.5)
# samples = number of samples, factor - ratio of samples
```

```python
plt.scatter(X[:,0], X[:, 1], c=y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
print(X_train.shape, X_test.shape)
```

### Classification using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

### Classification using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
# this may potentially collapse in case of close circles
```

### Classification using NeuralNetwork

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(solver='lbfgs',
                   alpha=1e-5,
                   hidden_layer_sizes=(6),
                   random_state=100,
                   activation='logistic',
                   max_iter=200) # alpha -> learning rate
mlp.fit(X_train, y_train)
mlp.score(X_test, y_test)
```
