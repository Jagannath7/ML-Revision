```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
img = cv2.imread('./img/landscape.jpg')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
print(img.shape)
```

```python
# kernel that will sharpen the image
K = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])
print(K.shape)
# a kernel with each element 0.1 can be used to blur the image
```

```python
out_row = img.shape[0] - K.shape[0] + 1
out_col = img.shape[1] - K.shape[1] + 1

out_img = np.zeros((out_row, out_col))
print(out_img.shape)
```

```python
for r in range(out_row):
    for c in range(out_col):
        img_patch = img[r:r+3, c:c+3]
        output = img_patch * K
        out_img[r,c] = output.sum() if out.sum()>0 else 0
```
